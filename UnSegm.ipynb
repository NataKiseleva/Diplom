```python
import pandas as pd
```

```python
df_k = pd.DataFrame(columns = ['ficha', 'weight', 'example', 'ist','types'])

for line in kitche:
  if line:
    if '\t' in line:
      pass
    else:
      line = line.replace('	', '\t', 1)
    if '\t' in line:
      print(line.split('\t'))
      if len(line.split('\t')) == 5:
        ficha, weight, example, ist, types = line.split('\t')
    d = {'ficha': ficha, 'weight': weight, 'example': example, 'ist': ist,'types': types}
    df_k = df_k.append(d, ignore_index=True)
```

```python
def l_to_l (df_1, df_2):
  f_l = list()
  for text in df_1['ficha']:
    for text2 in df_2['ficha']:
      if text == text2:
        f_l.append(text)
  return f_l
```

```python
def l_to_many (df_1, se):
  l_l_l = list()
  for k in range(0, len(se)):
    for text in df_1['ficha']:
      if se[k] == text:
        l_l_l.append(text)
  return l_l_l
```
